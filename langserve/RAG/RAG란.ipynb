```python
from langchain_upstage import ChatUpstage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

prompt = ChatPromptTemplate.from_messages([
    ("system", "한국어로 답변하세요"),
    ("user", "{user_input}")
])

llm = ChatUpstage()
chain = prompt | llm | StrOutputParser()

chain.invoke({"user_input": "아가멤논은 제우스가 보낸 꿈에서 어떤 장면을 보았나요?"})
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300, # 청크 사이즈를 조절
    chunk_overlap=50, # 청크 사이의 겹치는 부분의 크기를 조절
    length_function=len, # 문자열 길이를 계산하는 함수 지정
)
# documents 객체는 langchain에서 기본적으로 쓰는 객체
text_documents = text_splitter.create_documents([text])
print(text_documents[0])
print("---"*20)
print(text_documents[1])
```

```python
from langchain_upstage import UpstageEmbeddings

# 임베딩 모델 호출
passge_embedding = UpstageEmbeddings(model="solar-embedding-1-large-passage")
# documents 임베딩
embed_texts = passge_embedding.embed_documents([text])
embed_texts[0] # Text를 임베딩 한 결과값
```

```python
from langchain_chroma import Chroma

db = Chroma.from_documents(
    documents=text_documents, embedding=passge_embedding, collection_name="my_db"
)

db.get()
```

```python
db_retriever = db.as_retriever()

docs = db_retriever.invoke("아가멤논이 꾼 꿈은?")
docs
```

```python
from langchain.retrievers import EnsembleRetriever
from langchain_community.retrievers import BM25Retriever

bm25_retriever = BM25Retriever.from_texts(text)
bm25_retriever.k = 2

ensemble_retriever = EnsembleRetriever(
    retrievers=[db_retriever, bm25_retriever], weights=[0.5, 0.5]
)

docs = ensemble_retriever.invoke("아가멤논이 꾼 꿈은?")
docs
```
