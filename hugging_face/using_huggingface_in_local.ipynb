```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(
    base_url="http://localhost:1234/v1",
    api_key="lm-studio",
    model="koesn/SOLAR-10.7B-Instruct-v1.0-uncensored-GGUF",
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],  # 스트리밍 콜백
)

prompt = PromptTemplate.from_template(
    """You are a helpful, smart, kind, and efficient AI assistant. You always fulfill the user's requests to the best of your ability.
    You must generate an answer in Korean.

# QUESTION:
{question}

#Answer: """
)

chain = prompt | llm | StrOutputParser()
response = chain.invoke({"question": "한국의 수도는 어디야?"})
```

## LM Studio를 사용하여 로컬에서 LLM 작동시키기
코드만으로 할 부분이 아니라서 주석으로 적어둠

1. 먼저 LM Studio를 다운로드 받음
https://lmstudio.ai/

2. lm studio 설치 후 Search로 들어가서 원하는 huggingface 모델을 검색함
ex) heegyu/EEVE-Korean-Instruct-10.8B-v1.0-GGUF 와 가팅 모델명으로 검색해야함

!callout : 무조건 gguf라고 적힌 부분을 다운로드해야함

3. 이후 ai chat에 들어가서 윗 부분에 모델 설정 다운로드 받은 모델 설정
작동되는 지 확인 (chatgpt와 동일하다고 생각하고 질문해보기)

4. 왼쪽 측면바에 local server에 들어감

5. 설정 건드리지 말고 start server버튼 누르고 다시 모델 선택함

6. example 부분에서 chat(python) 부분 확인

7. 위에 코드에서 base_url, api_key, model을 바꾸어 주면됨 

<코드>

	from openai import OpenAI

	client = OpenAI(base_url="#http://localhost:1234/v1", api_key="#lm-studio")

	completion = client.chat.completions.create(
	model="#heegyu/EEVE-Korean-Instruct-10.8B-v1.0-GGUF",
	messages=[
	{"role": "system", "content": "Always answer in rhymes."},
	{"role": "user", "content": "Introduce yourself."}
	],
	temperature=0.7,
	)
	print(completion.choices[0].message)
